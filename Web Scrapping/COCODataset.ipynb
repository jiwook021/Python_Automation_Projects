```python
url ='https://cocodataset.org/#explore'
drivers.get(url)
soup = BeautifulSoup(drivers.page_source, 'html.parser')
```

```python

table_title = []
for i in range(len(soup.find("div", id ="exploreIconPanel").find_all("img"))):
    table_title.append(soup.find("div", id ="exploreIconPanel").find_all("img")[i]['title'])

pd.set_option('display.max_rows', 100)
df = pd.DataFrame(table_title,columns=['Contents'])
df
```
