```python
table_title = []
for i in range(len(soup.find('table').find_all("tr")[0].find_all("strong"))):
    table_title.append(soup.find('table').find_all("tr")[0].find_all("strong")[i].find("span").text)

for i in range(1,11):
    team = []
    team.append(i)
    team.append(soup.find('table').find_all("tr")[i].find('span').text)

    for k in range(1,len(soup.find('table').find_all("tr")[i].find_all('td'))):
        team.append(soup.find('table').find_all("tr")[i].find_all('td')[k].text)
    dicts = {}
    keys = table_title
    for j in range(len(keys)):
        dicts[keys[j]] = team[j]
    print(dicts)
    df = df.append(dicts,ignore_index=True)
```
